```python
import pandas as pd
url = 'https://raw.githubusercontent.com/krittikaiitb/KSP2021-selection/main/CometData.csv'
df = pd.read_csv(url)
print(df.head(5))
```

We assume that the Earth revolves around the Sun in a circular orbit of radius $1$ AU.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10, 8), dpi=80)
plt.polar(df['Angle between comet and sun(radians)'].values,df['Distance from earth(AU)'].values)
```

```python
#print(df.describe())
plt.plot(df['# Time(in years)'].values,df['Distance from earth(AU)'].values)
```

Note that the mean radial distance from the Earth increases with time. The oscillations are due to the revolution of the Earth itself around the sun, leading to apparent retrograde motions.

It is now necessary to convert the measurements to the sun frame. The arrays arrnDist and arrnAng represent the standard solar distance and
the Earth(at time 0)-Sun-Comet angles at all times. This data can then be fit with the equation of a standard conic.

```python
arrDist = np.array(df['Distance from earth(AU)'].copy())
arrAng = np.array(df['Angle between comet and sun(radians)'].copy())
arrTime = np.array(df['# Time(in years)'].copy())
type(arrDist)
arrnDist = np.sqrt(arrDist**2 + 1 - 2*arrDist*np.cos(arrAng))
arrnAng=[0 for i in range(0,len(arrAng))]

for i in range(0,len(arrAng)):
    if arrAng[i]>= np.pi/2 or arrAng[i]<= -np.pi/2:
        arrnAng[i] = 2*np.pi*arrTime[i] + np.pi - np.arcsin(arrDist[i]*np.sin(arrAng[i])/arrnDist[i])
    else:
        arrnAng[i] = 2*np.pi*arrTime[i] + np.arcsin(arrDist[i]*np.sin(arrAng[i])/arrnDist[i])

print(arrnDist[:5:1])
print(arrnAng[:5:1])
```

The conversion is as follows:

 $\sin\theta_{esc} = \frac{d_{ec}}{d_{sc}}\sin\theta_{ces}$
and
$d_{sc} = d_{ec}^2+d_{es}^2-2d_{ec}d_{es}\cos\theta_{ces}$

following the sine and cosine rule, where $c,e,s$ stand for comet, earth, sun respectively.

```python
plt.plot(df['# Time(in years)'].values,arrnDist)
```

```python
plt.figure(figsize=(10, 8), dpi=80)
plt.polar(arrnAng,arrnDist, color='orange', marker='o', markersize=5)
```

The radial distance was increasing with time as shown in the last plot, hence the trajectory starts from the interior of the figure and moves outwards.

```python
from scipy.optimize import curve_fit as cf
def orbit(theta, a, e, tilt):
    return a*(1-e**2)/(1+e*np.cos(theta-tilt))

p_opt, p_cov = cf(orbit,arrnAng,arrnDist)
```

The orbit function fits a standard conic of the form
$r(\theta) = \frac{r_0}{1+e\cos(\theta-\theta_0)}$
where $r_0 = a(1-e^2)$ is the semi-latus rectum, $e$ is the eccentricity and $\theta_0$ is the tilt of the trajectory.

```python
plt.figure(figsize=(10, 8), dpi=80)
plt.polar(arrnAng,arrnDist, color='orange', label='original data', marker='o', markersize=5)
plt.polar(arrnAng,orbit(arrnAng,*p_opt),label='Best Fit',color='g')
plt.legend(loc="lower center", bbox_to_anchor=(0.5, 1.15), ncol=2)
plt.title('Trajectory of comet around sun at origin')
```

We see that the Best fit actually fits the data almost perfectly. There are a couple outliers, but it is assumed that they fall under measurement errors, since these values are sufficiently close to the original trajectory.

```python
print(p_opt)
```

Note that p_opt is an array containing the values [a,e,tilt].

We find that the trajectory is an ellipse with semi major axis ($a$) $= 17.2288319$ AU, eccentricity($e$) $= 0.91373424 < 1$
and tilt angle $= 2.09474265$ rad.

$a(1-e) = 1.48625827$ AU is the distance of closest approach (perihelion) of the comet, $a(1+e) = 32.97140552$ AU being 
the aphelion of the comet.

From Newton-Kepler's laws, $a^3/T^2$ ($AU^3/year^2$) is constant for a celestial object orbiting the sun, so this ratio can be compared with the Earth, which has a ratio of 1. The time period of the comet is thus
T(comet) $= 71.51279470$ years

The calculations below show these data with greater precision.

```python
a = p_opt[0]
e = p_opt[1]
tilt = p_opt[2]

print('Eccentricity', e)
print()
print('Perihelion', a*(1-e), 'AU')
print()
print('Aphelion', a*(1+e), 'AU')
print()
print('Time Period', np.sqrt(a**3), 'years')
```

```python
thetaarr = np.array([i/1000 for i in range(0,2000+1000*int(np.floor(2*np.pi)),1)])
#print(thetaarr)
rarr = orbit(thetaarr, a, e, tilt)
rE = np.array([1 for i in range(0,len(thetaarr),1)])
rPluto = np.array([34.6 for i in range(0,len(thetaarr),1)])
```

```python
plt.figure(figsize=(10, 8), dpi=80)
plt.polar(arrnAng,arrnDist, color='orange', label='original data', marker='o', markersize=5)
#plt.polar(thetaarr, rarr, color='black', label='Comet orbit')
plt.polar(thetaarr,orbit(thetaarr,*p_opt),label='Comet orbit',color='g')
plt.polar(thetaarr, rE, color='blue', label='Earth orbit')
plt.polar(thetaarr, rPluto, color='brown', label='Pluto orbit')
plt.legend(loc="lower center", bbox_to_anchor=(0.5, 1.15), ncol=2)
plt.grid(False)
```

Note that Pluto is at an average distance of 34.6 AU from the sun. This shows that the comet belongs to the Solar System. In reality, Pluto goes around in an elliptical orbit which varies from 29.7 AU to 39.5 AU.
